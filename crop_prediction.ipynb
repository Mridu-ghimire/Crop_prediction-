```python
import pandas as pd

data = {
    'temperature': [20, 25, 30, 22, 28],
    'rainfall': [200, 150, 100, 180, 120],
    'crop': ['Rice', 'Wheat', 'Maize', 'Rice', 'Maize']
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X = df[['temperature', 'rainfall']]
y = df['crop']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

print("Model trained ✅")
```

```python
prediction = model.predict([[27, 130]])
print("Predicted Crop:", prediction[0])
```

```python
import pandas as pd

df = pd.read_csv("Crop_recommendation.csv")
df.head()
```

```python
X = df.drop('label', axis=1)
y = df['label']

print(X.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model trained successfully ✅")
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
prediction = model.predict([[90, 40, 40, 25, 80, 6.5, 200]])
print("Recommended Crop:", prediction[0])
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

dt_pred = dt_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

rf_acc = accuracy  # from previous model
dt_acc = accuracy_score(y_test, dt_pred)
knn_acc = accuracy_score(y_test, knn_pred)

print("Random Forest:", rf_acc)
print("Decision Tree:", dt_acc)
print("KNN:", knn_acc)
```
